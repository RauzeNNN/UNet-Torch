## Onnx Export

```python
# Some standard imports
import io
import numpy as np

from torch import nn
import torch.utils.model_zoo as model_zoo
import torch.onnx
```

```python
from Model import UNet, UNet_multitask, UNet_attention, UNet_fourier1, UNet_fourier1_2, UNet_BS
```

```python
model_path = "results/FTMT32out_exp5_35/epoch34.pt"

model = UNet_fourier1(1, 1, 32, True)
model.load_state_dict(torch.load(model_path), strict=False)
model.eval()
```

```python
# Input to the model
batch_size = 1
x = torch.randn(batch_size, 1, 512, 512, requires_grad=True)
model_out = model(x)

# Export the model
torch.onnx.export(model,               # model being run
                  x,                         # model input (or a tuple for multiple inputs)
                  "fourier.onnx",   # where to save the model (can be a file or file-like object)
                  export_params=True,        # store the trained parameter weights inside the model file
                  opset_version=11,          # the ONNX version to export the model to
                  do_constant_folding=True,  # whether to execute constant folding for optimization
                  input_names = ['input'],   # the model's input names
                  output_names = ['output1',"output2"], # the model's output names
                  dynamic_axes={'input' : {0 : 'batch_size'},    # variable length axes
                                'output' : {0 : 'batch_size',1:"batch_size"}})
```
